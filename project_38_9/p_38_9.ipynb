# ***Анализ поведения пользователей в продукте 'Мобильная игра Quiz Freeze'***

# ***Решаемые задачи:***
## Зависит ли вероятность оплаты от выбранного пользователем уровня сложности?
## Существует ли разница во времени между событиями регистрации и оплаты для разных групп пользователей с разным уровнем сложности?

## ***0. Общие характеристики сегментов пользователей по уровням сложности выбираемых бесплатных пакетов вопросов:***

### Структура пользователей:
Всего 8342 человека, совершивших ***выбор уровня*** (=100%). Из них по уровням сложности вопросов:
* medium -   4645 человек, или 55.7%
* easy -     2448 человек, или 29.3%
* hard -     1249 человек, или 15.0%

Более половины пользователей выбирают уровень ***medium***, из оставшихся порядка 1/3 предпочитают сложный уровень ***hard***, 2/3 - более легкий уровень ***easy***

### Структура выручки:
Всего ***оплата*** 177175 руб.=100%. Из них поступило от пользователей, которые ранее выбрали уровни вопросов:
* medium -   106125 руб., или 59.9%
* easy -     21724 руб., или 12.3%
* hard -     49327 руб., или 27.8%

*Почти 60% выручки обеспечено сегментом пользователей, выбравших уровень ***medium*** бесплатных вопросов, вклад пользователей сегментов hard и easy в выручку - зеркальный относительно распределения их количественного распределения. Высокая доля сегмента ***hard*** (почти 28%), слабая - сегмента ***easy*** (порядка 12%)*

### ***Выводы*** относительно структурных характеристик:

Базовый пользовательский сегмент - ***medium***. Он решающим образом определяет динамику всех показателей, его мониторинг - безусловная постоянная задача.

Блок вопросов ***hard*** очень удачный. При минимуме пользователей он обеспечивает значимую часть выручки.

Блок ***easy*** требует анализа и доработки в связи со слабой долей в формировании выручки.

## ***1. Как зависит вероятность оплаты от выбранного пользователем уровня сложности? + Какой он, средний платеж по сегментам пользователей?***

В целом в приложении ***производят оплату*** 19.18% пользователей, которые ранее выбрали уровень сложности бесплатного пакета вопросов. В зависимости от уровня сложности вопросов, эта доля составляет:
* medium -   20.86%
* easy -     7.72%
* hard -     35.39% 

В целом в приложении ***средний платеж*** составляет 110.7 руб. В зависимости от уровня сложности вопросов, он равен:
* medium -   109.5 руб.
* easy -     115.0 руб. 
* hard -     111.6 руб. 

#### Без статистики в динамике нельзя делать выводы об эффективности ***блоков medium, easy, hard*** с точки зрения выручки. 
### Очевидные ***выводы*** в моменте: 

***Блок medium*** - он действительно определяет общий уровень монетизации игры. Его динамику надо отслеживать системно. По факту, почти 21% пользователей переходит к оплате, но не ясно, насколько это хорошо в сравнении с 7.72% easy и 35.39% hard. Что ясно, это относительно небольшой средний платеж по сегменту. Насколько такое поведение пользователя (уровень мотивации и невысокий чек) ожидаемо для разработчиков? 

***Блок easy*** слабый. Необходимо понять, почему он мотивирует к продолжению игры менее 8% выбравших этот пакет вопросов, средний платеж которых, однако, достаточно высокий. Это важно, поскольку (1) изначально этот уровень предпочитает более четверти пользователей и (2) конверсия из него в оплату настолько низкая, что утягивает за собой и итоговый по всему продукту уровень.

***Блок hard*** очень полезный. Более 35% пользователей после ознакомления с бесплатным пакетом вопросов переходят к оплате, и средний платеж достаточно высокий. 

## ***2. Существует ли разница во времени между событиями регистрации и оплаты для разных групп пользователей с разным уровнем сложности?***

По продукту в целом, для пользователей, выбравших уровень сложности бесплатного пакета вопросов, ***среднее время между регистрацией и оплатой*** составляет 4 days 01:01:56.595000. В зависимости от выбранного уровня, этот показатель
* medium -   4 days 06:12:06.576883384
* easy -     3 days 22:10:23.211640211
* hard -     3 days 14:55:19.257918552

По продукту в целом, для пользователей, выбравших уровень сложности бесплатного пакета вопросов, ***среднее время между выбором уровня сложности и оплатой*** составляет 3 days 17:52:17.719375. В зависимости от выбранного уровня, этот показатель
* medium -   3 days 23:14:13.165118679
* easy -     3 days 14:58:52.941798941
* hard -     3 days 07:20:41.420814479

### Основные ***выводы***:  

***Блок hard***. Выбирающие этот уровень бесплатного пакета вопросов характеризуются не только высокой мотивацией к продолжению игры (оплате) и высоким средним платежом, но и высокой реактивностью. Среднее время от регистрации или выбора уровня до оплаты минимальное по трем сегментам: 3 days 14:55:19 и 3 days 07:20:41, соответственно. Перспективный сегмент, эффективный блок вопросов. Выписать разработчикам премию и улучшать дальше. 

***Блок easy***. Пользователи, выбравшие бесплатный пакет вопросов этого уровня сложности, также принимают решение об оплате достаточно быстро - в среднем, через 3 days 22:10:23 после регистрации на сайте и через 3 days 14:58:53 после выбора уровня вопросов. При доработке этого блока вопросов важно учитывать, что тем 7.72% пользователям, которые решили продолжить игру, не понадобилось много времени для совершения относительно большого платежа. Возможно, нужна более четкая фокусировка уровень-пользователь?  

***Блок medium***. Черный ящик. Достаточны ли 20.86% конверсии из вопросов в оплату - не факт. Факт - самый низкий по сегментам средний платеж и самый большой лаг до оплаты: 4 days 06:12:07 от регистрации, 3 days 23:14:13 от выбора уровня вопросов. Пользователь сегмента - не самый мотивированный, не самый щедрый, не самый быстрый игрок. Он неявно отказался от выбора на входе (не easy/hard, а medium), и в дальнейшем ведет себя неактивно. Это самый большой сегмент, от него зависит многое. Надо что-то делать. Возможно, как-то сегментировать/премировать этот блок вопросов, побуждая пользователя к более активным действиям?
